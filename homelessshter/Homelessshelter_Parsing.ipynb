```python
import pandas as pd
import numpy as np
import os
import requests
```

```python
def is_downloadable(url):
    """
    Does the url contain a downloadable resource
    """
    h = requests.head(url, allow_redirects=True)
    header = h.headers
    content_type = header.get('content-type')
    if 'text' in content_type.lower():
        return False
    if 'html' in content_type.lower():
        return False
    return True

print is_downloadable('https://www.homelessshelterdirectory.org/cgi-bin/id/city.cgi?city=New%20York&state=NY')
# >> False
print is_downloadable('https://www.shelterlistings.org/city/new_york-ny.html')
# >> True
print is_downloadable('https://data.cityofnewyork.us/Social-Services/Directory-Of-Homebase-Locations/ntcm-2w4k/data')
```
